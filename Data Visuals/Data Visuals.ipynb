```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
db_path = '/Users/apple/DSI/team_project_2/data/sql/Canada_Housing_Prices_2005_2024.db'
conn = sqlite3.connect(db_path)

# Load the data from the key tables
housing_price_index = pd.read_sql_query("SELECT * FROM Housing_Price_Index", conn)
aggregate_data = pd.read_sql_query("SELECT * FROM AGGREGATE", conn)
interest_rate_data = pd.read_sql_query("SELECT * FROM Bank_of_Canada_Interest_Rate", conn)

# Convert the Date columns to datetime format
housing_price_index['Date'] = pd.to_datetime(housing_price_index['Date'])
aggregate_data['Date'] = pd.to_datetime(aggregate_data['Date'])
interest_rate_data['Date'] = pd.to_datetime(interest_rate_data['Date'])

# Plot the trend of the Composite Housing Price Index over time
plt.figure(figsize=(12, 6))
plt.plot(aggregate_data['Date'], aggregate_data['Composite_HPI_SA'], label='Composite HPI (Seasonally Adjusted)')
plt.title('Trend of Composite Housing Price Index (2005 - 2024)')
plt.xlabel('Year')
plt.ylabel('Composite HPI')
plt.grid(True)
plt.legend()
plt.show()
```

The graph indicates a general upward trend in the housing prices across Canada over the last two decades.
There are periods of accelerated growth, particularly noticeable around 2016-2017 and again post-2020, which could be due to various economic factors, including changes in demand, supply constraints, and possibly the impact of economic stimuli during the COVID-19 pandemic.
The consistent increase suggests that housing has become significantly more expensive over time, reflecting market conditions such as low interest rates, population growth, and possibly speculative investment in real estate.

```python
# Plot the trend of different housing categories over time
plt.figure(figsize=(14, 7))
plt.plot(aggregate_data['Date'], aggregate_data['Single_Family_HPI_SA'], label='Single Family HPI')
plt.plot(aggregate_data['Date'], aggregate_data['One_Storey_HPI_SA'], label='One Storey HPI')
plt.plot(aggregate_data['Date'], aggregate_data['Two_Storey_HPI_SA'], label='Two Storey HPI')
plt.plot(aggregate_data['Date'], aggregate_data['Townhouse_HPI_SA'], label='Townhouse HPI')
plt.plot(aggregate_data['Date'], aggregate_data['Apartment_HPI_SA'], label='Apartment HPI')
plt.title('Comparison of Housing Prices Across Different Categories (2005 - 2024)')
plt.xlabel('Year')
plt.ylabel('Housing Price Index')
plt.grid(True)
plt.legend()
plt.show()

```

All categories exhibit an upward trend, similar to the overall Composite HPI, indicating that prices across all housing types have increased.
The Two Storey and Single Family homes show the highest price indices, suggesting that these types of properties have appreciated the most over the period. This might be due to higher demand for larger family homes.
Apartment and Townhouse categories have a lower but still increasing trend, which could indicate that these property types, while becoming more expensive, may be less in demand or more readily available compared to detached homes.
The differences in growth rates among these categories might reflect changes in buyer preferences, availability of land for construction, and economic factors influencing the type of housing people can afford.

```python

# Plot Interest Rate vs. Composite HPI
fig, ax1 = plt.subplots(figsize=(14, 7))

# Plot the Composite HPI on the first y-axis
ax1.plot(aggregate_data['Date'], aggregate_data['Composite_HPI_SA'], color='tab:blue', label='Composite HPI')
ax1.set_xlabel('Year')
ax1.set_ylabel('Composite HPI', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a second y-axis to plot the Interest Rate
ax2 = ax1.twinx()
ax2.plot(interest_rate_data['Date'], interest_rate_data['Interest Rate'], color='tab:red', label='Interest Rate')
ax2.set_ylabel('Interest Rate (%)', color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Add title and grid
plt.title('Interest Rate vs. Composite Housing Price Index (2005 - 2024)')
fig.tight_layout()
plt.grid(True)
plt.show()








```

The graph suggests an inverse relationship between interest rates and housing prices. As interest rates decrease, housing prices tend to increase.
During periods of low interest rates, such as after the 2008 financial crisis and again post-2020, housing prices saw significant increases. Low borrowing costs typically encourage home buying, driving up demand and, consequently, prices.
Conversely, any rise in interest rates might have been expected to slow down the price increase, but the graph indicates that housing prices continued to rise despite occasional rate hikes. This could reflect underlying demand strength, low inventory, or speculative behavior in the housing market.
The overall trend suggests that while interest rates do influence housing prices, other factors like supply-demand dynamics, government policies, and economic conditions also play crucial roles.
