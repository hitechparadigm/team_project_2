### Housing prices dataset: by provinces and urban areas

```python
import pandas as pd

# Load the spreadsheet
file_path = '../data/raw/Canada_Housing_Prices_2005_2024.xlsx'
spreadsheet = pd.ExcelFile(file_path)

# Display the names of the sheets
sheet_names = spreadsheet.sheet_names
sheet_names
```

```python
import sqlite3

# Create a new SQLite database
db_path = '../data/sql/Canada_Housing_Prices_2005_2024.db'
conn = sqlite3.connect(db_path)

# Function to write each sheet to the SQLite database
def write_sheet_to_db(sheet_name):
    df = spreadsheet.parse(sheet_name)
    df.to_sql(sheet_name, conn, if_exists='replace', index=False)

# Write each sheet to the database
for sheet in sheet_names:
    write_sheet_to_db(sheet)

# Close the connection
conn.close()

# Confirm the database creation
db_path
```

```python
# Load the data from the 'AGGREGATE' sheet
conn = sqlite3.connect(db_path)
aggregate_df = pd.read_sql_query("SELECT * FROM AGGREGATE", conn)
conn.close()

# Display the data from 'AGGREGATE' sheet

aggregate_df.head()
```

```python
# Create a dataset for Single Family House Prices

import pandas as pd
import sqlite3
import os

# Connect to the SQLite database
db_path = '../data/sql/Canada_Housing_Prices_2005_2024.db'
conn = sqlite3.connect(db_path)

# Fetch all table names from the database
query = "SELECT name FROM sqlite_master WHERE type='table';"
table_names = pd.read_sql_query(query, conn)['name'].tolist()

# Print the fetched table names for debugging
print("Fetched table names:", table_names)

# Initialize the main data frame
main_df = None

# Read data from each table and extract relevant columns
for table in table_names:
    df = pd.read_sql_query(f"SELECT Date, Single_Family_Benchmark_SA FROM {table}", conn)
    if not df.empty:
        df.rename(columns={"Single_Family_Benchmark_SA": table}, inplace=True)
        if main_df is None:
            main_df = df
        else:
            main_df = pd.merge(main_df, df, on="Date", how="outer")
    else:
        print(f"Warning: Table '{table}' is empty and will not be included.")

# Close the connection
conn.close()

# Save the combined dataset to a new file
output_path = '../data/processed/Single_Family_Benchmark_SA_All_Geographies.csv'
main_df.to_csv(output_path, index=False)

# Display the first few rows of the combined dataset
print("Combined DataFrame:")
print(main_df.head())

```

```python
# Create a dataset for Apartment Prices
# Not all the tables have Apartnment Prices. Extract only from the tables that have this data

import pandas as pd
import sqlite3
import os

# Connect to the SQLite database
db_path = '../data/sql/Canada_Housing_Prices_2005_2024.db'
conn = sqlite3.connect(db_path)

# Fetch all table names from the database
query = "SELECT name FROM sqlite_master WHERE type='table';"
table_names = pd.read_sql_query(query, conn)['name'].tolist()

# Print the fetched table names for debugging
print("Fetched table names:", table_names)

# Initialize the main data frame
main_df = None

# Function to check if a column exists in a table
def column_exists(table_name, column_name):
    query = f"PRAGMA table_info({table_name});"
    columns_info = pd.read_sql_query(query, conn)
    return column_name in columns_info['name'].values

# Read data from each table and extract relevant columns
for table in table_names:
    if column_exists(table, 'Apartment_Benchmark_SA'):
        df = pd.read_sql_query(f"SELECT Date, Apartment_Benchmark_SA FROM {table}", conn)
        if not df.empty:
            df.rename(columns={"Apartment_Benchmark_SA": table}, inplace=True)
            if main_df is None:
                main_df = df
            else:
                main_df = pd.merge(main_df, df, on="Date", how="outer")
    else:
        print(f"Warning: Table '{table}' does not have column 'Apartment_Benchmark_SA' and will not be included.")

# Close the connection
conn.close()

# Save the combined dataset to a new file
output_path = '../data/processed/Apartment_Benchmark_SA_All_Geographies.csv'
main_df.to_csv(output_path, index=False)

# Display the first few rows of the combined dataset
print("Combined DataFrame:")
print(main_df.head())
```

```python
import pandas as pd
import sqlite3
import os

# Connect to the SQLite database
db_path = '../data/sql/Canada_Housing_Prices_2005_2024.db'
conn = sqlite3.connect(db_path)

# Fetch all table names from the database
query = "SELECT name FROM sqlite_master WHERE type='table';"
table_names = pd.read_sql_query(query, conn)['name'].tolist()

# Print the fetched table names for debugging
print("Fetched table names:", table_names)

# Initialize a list to hold data frames
data_frames = []

# Read data from each table and extract all columns
for table in table_names:
    try:
        df = pd.read_sql_query(f"SELECT * FROM {table}", conn)
        if not df.empty:
            # Prefix column names with table name to avoid collisions
            df = df.rename(columns={col: f"{table}_{col}" for col in df.columns if col != 'Date'})
            data_frames.append(df)
        else:
            print(f"Warning: Table '{table}' is empty and will not be included.")
    except Exception as e:
        print(f"Error processing table '{table}': {e}")

# Close the connection
conn.close()

# Concatenate all data frames along columns, aligning on 'Date'
if data_frames:
    main_df = pd.concat(data_frames, axis=1, join='outer')
    # Ensure that 'Date' column is not duplicated
    main_df = main_df.loc[:, ~main_df.columns.duplicated()]

    # Save the combined dataset to a new file
    output_path = '../data/processed/Housing_Prices_All_Data.csv'
    main_df.to_csv(output_path, index=False)
    
    # Display the first few rows of the combined dataset
    print("Combined DataFrame:")
    print(main_df.head())
else:
    print("No data to save. All tables were empty or there was an error reading the data.")
```
