```python
import pandas as pd

# Load the spreadsheet
file_path = '../data/raw/Canada_Housing_Prices_2005_2024.xlsx'
spreadsheet = pd.ExcelFile(file_path)

# Display the names of the sheets
sheet_names = spreadsheet.sheet_names
sheet_names
```

```python
import sqlite3

# Create a new SQLite database
db_path = '../data/sql/Canada_Housing_Prices_2005_2024.db'
conn = sqlite3.connect(db_path)

# Function to write each sheet to the SQLite database
def write_sheet_to_db(sheet_name):
    df = spreadsheet.parse(sheet_name)
    df.to_sql(sheet_name, conn, if_exists='replace', index=False)

# Write each sheet to the database
for sheet in sheet_names:
    write_sheet_to_db(sheet)

# Close the connection
conn.close()

# Confirm the database creation
db_path
```

```python
# Load the data from the 'AGGREGATE' sheet
conn = sqlite3.connect(db_path)
aggregate_df = pd.read_sql_query("SELECT * FROM AGGREGATE", conn)
conn.close()

# Display the data from 'AGGREGATE' sheet

aggregate_df.head()
```
